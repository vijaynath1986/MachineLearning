```python
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import pandas as pd
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
dat = sns.load_dataset("tips")
```

```python
sns.get_data_home() #copy the dataset to this location inorder to load the dataset
```

```python
dat
```

```python
ssObj = StandardScaler()
```

```python
tips_numerical = dat[["total_bill", "tip", "size"]]
tips_numerical
```

```python
tips_numerical.describe()
```

```python
ssObj.fit(tips_numerical)
```

```python
X_tr = ssObj.transform(tips_numerical)
pd.DataFrame(X_tr).describe()
```

```python
ssObj.inverse_transform(X_tr)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mmObj = MinMaxScaler(feature_range = (5, 10))
```

```python
mmObj.fit_transform(tips_numerical)
```

```python
pd.DataFrame(mmObj.fit_transform(tips_numerical), columns= tips_numerical.columns).describe()
```
