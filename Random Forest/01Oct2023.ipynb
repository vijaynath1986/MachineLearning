```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
dat = pd.read_csv("../DataSets/pacific.csv")
dat_presever = dat.copy()
```

```python
dat.drop(labels=  ["ID", "Name", "Date", "Event", "Latitude", "Longitude"], axis = 1, inplace=True)
```

```python
dat.info()
```

```python
dat["Status"].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oeObj = OrdinalEncoder()
dat["status_n"] = oeObj.fit_transform(dat[["Status"]])
```

```python
oeObj.feature_names_in_
```

```python
dat.drop(labels =["Status"], inplace=True, axis=1)
```

```python
sns.heatmap(dat.corr())
```

```python
dat
```

```python
X = dat.drop("status_n", axis=1)
y = dat["status_n"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainX, testX, trainy, testy = train_test_split(X, y, test_size=0.3)
```

```python
print(trainX.shape, testX.shape, trainy.shape, testy.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfcObj = RandomForestClassifier(n_estimators=200)
```

```python
rfcObj.fit(trainX, trainy)
```

```python
predy = rfcObj.predict(testX)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import balanced_accuracy_score
```

```python
confusion_matrix(testy, predy)
```

```python
accuracy_score(testy, predy)
```

```python
balanced_accuracy_score(testy, predy)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(testy, predy))
```
