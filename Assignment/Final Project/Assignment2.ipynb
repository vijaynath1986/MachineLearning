```python
df = pd.read_csv("542_proj_dataset_2_v1.0/Project_Data_2.csv", thousands= ',')
df.rename(columns={"Sales of Wheat in tons": 'Country'}, inplace=True)
#df.set_index("Country", inplace=True)
df.head()
```

```python
df_num_only = df.drop("Country", axis=1)
plt.figure(figsize=(10,10))
sns.heatmap(df_num_only.corr(), annot=True)
```

```python
sns.pairplot(df_num_only)
```

```python
from sklearn.preprocessing import StandardScaler 

ssobj = StandardScaler()
df_scaled = ssobj.fit_transform(df_num_only)
df_scaled = pd.DataFrame(df_scaled)
df_scaled.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
df_pca = pca.fit_transform(df_scaled)
df_pca
```

```python
df_pca = pd.DataFrame(df_pca)
df_pca.head()
```

```python
pca.explained_variance_ratio_*100
```

```python
plt.plot(range(0, len(pca.explained_variance_ratio_)), pca.explained_variance_ratio_, 'go-', linewidth = 1, label=pca.explained_variance_ratio_ )
```

```python
df_new = df_pca[[0,1,2]]
df_new
```

```python
df_new_country = df_new.assign(Country = df["Country"])
df_new_country.head()
```
