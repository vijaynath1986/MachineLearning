```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler 

ssobj = StandardScaler()
df_scaled = ssobj.fit_transform(df)

df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df_scaled.head()
```

```python
df_scaled.set_index(df.index, inplace=True)
df_scaled.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
df_pca = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(df_pca)
print(pca.explained_variance_ratio_*100)
df_pca.head()
```

```python
plt.plot(range(0, len(pca.explained_variance_ratio_)), pca.explained_variance_ratio_*100, 'go-', linewidth = 1, label=pca.explained_variance_ratio_ )
```

## 95% of variance is captured by using only 2 columns/new transformed features. 

```python
df_pca.set_index(df.index, inplace=True)
df_pca.head()
```

```python
sns.scatterplot(df_pca)
```

```python
df_pca[0]
```

```python
from sklearn.cluster import KMeans
x = df_pca
wcss = []
for each in range(1, 11):
    kmobj = KMeans(n_clusters=each)
    kmobj.fit_predict(x)
    wcss.append(kmobj.inertia_)
```

```python
print(wcss)
plt.plot(wcss, 'ro-')
plt.xlabel("clusters")
plt.ylabel("Sum of sq disctance between clusters")
```

```python
kmobj = KMeans(n_clusters=4)
kmobj.fit_predict(x)
print(kmobj.labels_)
```

```python
x["label"] = kmobj.labels_
```
