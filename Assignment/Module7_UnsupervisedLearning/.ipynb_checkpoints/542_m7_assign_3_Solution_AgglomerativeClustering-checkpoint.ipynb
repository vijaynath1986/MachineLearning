## 3. Use the 16-intermediate feature and perform an agglomerative clustering.
[ Hint: Refer to the agglomerative clustering module in scikit learn and set the number of clusters appropriately]

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clustering = AgglomerativeClustering(n_clusters=7)
```

```python
X = dat.drop(["animal_name", "class_type"], axis=1)
```

```python
clustering.fit(X)
```

```python
clustering.labels_
```

```python
clustered = clustering.labels_+1
```

```python
predicted_cluster = pd.DataFrame(clustered, columns=["predicted_cluster"])
predicted_cluster
```

```python
plt.figure(figsize=(20, 10))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

```python
dat = pd.concat([dat, predicted_cluster], axis=1)
```

```python
dat
```

```python
sns.lineplot(data = dat, x= "class_type", y = "predicted_cluster" )
```

```python
## Mean square error is not suitable for clustering problems. Confusion metrics gives better picture.
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confMatrix = confusion_matrix(dat["class_type"], dat["predicted_cluster"])
```
