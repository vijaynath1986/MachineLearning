```python
basket = dat_france.groupby(["InvoiceNo" , "Description"])["Quantity"].sum().unstack().fillna(0)
```

```python
basket.head(20)
```

```python
def encode_units(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
basket_set = basket.applymap(encode_units)
```

```python
basket_set.head(20)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
freq_itemset = apriori(basket_set, min_support=0.07, use_colnames=True)
```

```python
freq_itemset
```

```python
association_rules(freq_itemset, metric="confidence", min_threshold=0.5)
```

```python
association_rules(freq_itemset, metric="lift", min_threshold=1)
```
