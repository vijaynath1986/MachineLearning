## 1. We will use acoustic features to distinguish a male voice from female. Load the dataset from “voice.csv”, identify the target variable and do a one-hot encoding for the same. Split the dataset in train-test with 20% of the data kept aside for testing.
[Hint: Refer to LabelEncoder documentation in scikit-learn]

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohemodel = OneHotEncoder()
```

```python
oheResult = ohemodel.fit_transform(dat[["label"]])
```

```python
oheResult = pd.DataFrame(oheResult.toarray(), columns= ["label_male_0_female_1", "label_male_1_female_0"])
```

```python
oheResult.head()
```

```python
y = oheResult.drop(labels= ["label_male_1_female_0"], axis=1)
```

```python
x = dat.drop(labels="label", axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.2, random_state=85)
```

```python
y
```

```python
#Append the y with dat.
dat = pd.concat([dat, y], axis=1)
```

```python
dat
```

## 2. Fit a logistic regression model and measure the accuracy on the test set.
[Hint: Refer to Linear Models section in scikit-learn]

```python
from sklearn.linear_model import LogisticRegression
```

```python
logregModel = LogisticRegression()
```

```python
logregModel.fit(train_x, train_y)
```

```python
predicted_y = logregModel.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(test_y, predicted_y)
```

```python
logregModel.score(test_x, test_y) *100
```
