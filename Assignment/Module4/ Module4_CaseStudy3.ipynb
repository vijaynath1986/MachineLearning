```python
paid_record, not_paid_record
```

```python
dat_paid_record = dat[(dat["not.fully.paid"]==0)].sample(not_paid_record)
```

```python
dat_not_paid_record = dat[(dat["not.fully.paid"]==1)]
```

```python
dat_undersampled = pd.concat([dat_paid_record, dat_not_paid_record], axis=0)
```

```python
dat_undersampled
```

```python
x_us = dat_undersampled.drop(labels=["purpose", "not.fully.paid"], axis=1)
```

```python
y_us = dat_undersampled["not.fully.paid"]
```

```python
train_x_us, test_x_us, train_y_us, test_y_us = train_test_split(x_us, y_us, test_size=0.33, stratify=y_us)
```

```python
print(train_x_us.shape, test_x_us.shape, train_y_us.shape, test_y_us.shape)
```

```python
logRegModel_us = LogisticRegression()
```

```python
logRegModel_us.fit(train_x_us, train_y_us)
predicted_y_us = logRegModel_us.predict(test_x_us)
logRegModel_us.score(test_x_us, test_y_us)
```

```python
from sklearn.metrics import classification_report
```
