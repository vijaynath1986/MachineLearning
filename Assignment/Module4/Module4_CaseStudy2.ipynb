## 4) Fit a decision tree classifier and observe the accuracy.

```python
plt.figure(figsize=(20,15))
sns.heatmap(dat_filled.corr(), annot=True)
```

```python
y = dat_filled["outcome"]
```

```python
x = dat_filled.drop(labels= ["hospital_number", "outcome", "lesion_1", "lesion_2", "lesion_3", 
                             "nasogastric_reflux_ph", "peripheral_pulse"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=85)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtcModel = DecisionTreeClassifier()
```

```python
dtcModel.fit(train_x, train_y)
```

```python
predicted_y = dtcModel.predict(test_x)
```

```python
dtcModel.score(test_x, test_y)*100
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(test_y, predicted_y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfcModel = RandomForestClassifier()
```

```python
rfcModel.fit(train_x, train_y)
```

```python
rf_predicted_y = rfcModel.predict(test_x)
```

```python
rfcModel.score(test_x, test_y)*100
```

```python
confusion_matrix(test_y, rf_predicted_y)
```

## Accuracy imporved from Decision Tree classification to Random Forest Classification
