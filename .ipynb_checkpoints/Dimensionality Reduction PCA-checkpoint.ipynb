```python
df= pd.read_csv("trans_us.csv", thousands=',')
```

```python
df.info()
```

```python
df.columns[0]
```

```python
df.set_index(df.columns[0], inplace=True)
```

```python
corr = df.corr()
```

```python
corr
```

```python
sns.heatmap(corr)
```

```python
from sklearn.decomposition import PCA
```

```python
pcaObj = PCA(n_components=18)
```

```python
transformed_array = pcaObj.fit_transform(df)
```

```python
transformed_array
```

```python
transformed_df = pd.DataFrame(transformed_array)
```

```python
transformed_df
```

```python
pcaObj.explained_variance_
#This is eigen value
```

```python
pcaObj.explained_variance_ratio_*100
#This should how much % of value contribute to each column
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[20, 5])
plt.bar(df.columns, pcaObj.explained_variance_ratio_*100)
# This shows keeping first two PCA transformed column is good enough to get major accurate prediction.  
# Other columns contribute to negligible
```

```python
pcaObjNew = PCA(n_components=2)
transformed_array_new = pcaObjNew.fit_transform(df)
pcaObjNew.explained_variance_ratio_*100
```

```python
#divide dependend x and independent variable y
x = df[['1990', '1991']]
y = df['2007']
```

```python
# divide training and testing set
from sklearn.model_selection import train_test_split
```

```python
trainx, testx, trainy, testy = train_test_split(x, y, test_size=0.3, random_state=5)
```

```python
print(trainx.shape)
print(trainy.shape)
print(testx.shape)
print(testy.shape)
```

```python
transformed_trainx = pcaObjNew.fit_transform(trainx)
transformed_testx = pcaObjNew.fit_transform(testx)
```

```python
pcaObjNew.explained_variance_ratio_
```

```python
# Fitting Logistic Regression To the training set
from sklearn.linear_model import LinearRegression
```

```python
classifier = LinearRegression()
classifier.fit(transformed_trainx, trainy)
```

```python

# Predicting the test set result using
# predict function under LogisticRegression
predictedy = classifier.predict(transformed_testx)
```

```python
testy
```

```python
pd.DataFrame({"trainy":testy, "predicted y" :predictedy})
```
