```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../DataSets/WineQuality/winequality-red.csv', sep = ';')
data.head()
```

```python
plt.figure(figsize=(20,7))
sns.scatterplot(data = data, x= "fixed acidity", y = "volatile acidity", hue=data["quality"], palette= "bright")
```

```python
plt.figure(figsize=(20,7))
sns.scatterplot(data = data, x= "citric acid", y = "residual sugar", hue=data["quality"], palette= "bright")
```

```python
plt.figure(figsize=(20,7))
sns.scatterplot(data = data, x= "density", y = "alcohol", hue=data["quality"], palette= "bright")
```

```python
from sklearn.model_selection import train_test_split
columns = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", 
           "total sulfur dioxide", "density","pH", "sulphates", "alcohol"]
X = data[columns]
y = data["quality"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
DecisionTreeModel = DecisionTreeClassifier(criterion='gini', min_samples_split= 2)
DecisionTreeModel.fit(X_train, y_train)
DecisionTreeModel.score(X_test, y_test)
```

```python
from sklearn.model_selection import train_test_split

columns = ["alcohol"]
X = data[columns]
y = data["quality"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
DecisionTreeModel = DecisionTreeClassifier(criterion='gini', min_samples_split= 2)
DecisionTreeModel.fit(X_train, y_train)
DecisionTreeModel.score(X_test, y_test)
```
