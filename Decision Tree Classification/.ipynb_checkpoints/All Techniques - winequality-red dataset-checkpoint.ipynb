```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn import svm



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../DataSets/WineQuality/winequality-red.csv', sep = ';')
data.head()
```

```python
# Check for null, nan, object items
data.info()
```

```python
#I wanna see how each feature is correlated
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot=True)
```

```python
sns.pairplot(data, hue='quality', palette= 'bright')
## Palette - deep, muted, pastel, bright, dark, and colorblind
```

```python
# From above pairplots there is no correlation amoung the features. So lets try decision tree first and 
# comapre with linear regression later
```

## Data standardization

```python
# Standardize the features
from sklearn.preprocessing import StandardScaler
```

```python
X = data.iloc[:, :-1]
y = data["quality"]
```

```python
std = StandardScaler()
X_std = std.fit_transform(X)
```

```python
print(X_std)
```

## Train test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size= 0.3, random_state=5)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## DecisionTreeRegressor

```python
decsionTreeRegModel = DecisionTreeRegressor()
decsionTreeRegModel.fit(X_train, y_train)
decsionTreeRegModel.score(X_test, y_test)
```

## LinearRegression

```python
LinearRegModel = LinearRegression()
LinearRegModel.fit(X_train, y_train)
LinearRegModel.score(X_test, y_test)
```

## DecisionTreeClassifier

```python
DecisionTreeModel = DecisionTreeClassifier(criterion='gini', min_samples_split= 2)
DecisionTreeModel.fit(X_train, y_train)
DecisionTreeModel.score(X_test, y_test)
```

## PolynomialFeatures

```python
PolynomialFeaturesModel = PolynomialFeatures(degree=2)
LinearRegModel = LinearRegression()
LinearRegModel.fit(PolynomialFeaturesModel.fit_transform(X_train), y_train)
LinearRegModel.score(PolynomialFeaturesModel.fit_transform(X_test), y_test)
```

## RandomForestClassifier

```python
RandomModel = RandomForestClassifier(n_estimators = 100)
RandomModel.fit(X_train, y_train)
RandomModel.score(X_test, y_test)
#### THis result is better than anyother model so far, which is ~69%
```

## RandomForestRegressor

```python
RandomRegModel = RandomForestRegressor(n_estimators = 100)
RandomRegModel.fit(X_train, y_train)
RandomRegModel.score(X_test, y_test)
```

## SVM

```python
svmClassModel = svm.SVC()
svmClassModel.fit(X_train, y_train)
svmClassModel.score(X_test, y_test)
```

```python
## None of the above models gave satisfactory accuracy score so see the features contributing for information to predict
```

## PCA

```python
from sklearn.decomposition import PCA
pcaObj = PCA()
X_pca = pcaObj.fit_transform(X_std)
pd.DataFrame(X_pca)
pcaObj.explained_variance_
```

```python
pcaObj.explained_variance_ratio_
```

```python
sum([0.28173931, 0.1750827 , 0.1409585 , 0.11029387, 0.08720837,
       0.05996439, 0.05307193])
# First Seven Components contribute for 90% information so lets see, i believe after dimensionality reduction 
# accuracy comes down further
```

```python
pcaObj = PCA(n_components=7)
X_pca = pcaObj.fit_transform(X_std)
pd.DataFrame(X_pca)
pcaObj.explained_variance_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size= 0.3, random_state=5)
```

```python
RandomModel = RandomForestClassifier(n_estimators = 100)
RandomModel.fit(X_train, y_train)
RandomModel.score(X_test, y_test)
```

```python
## As expected, Same accuracy score as before PCA
```

## LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
ldaModel = LinearDiscriminantAnalysis()
ldaModel.fit(X_train, y_train)
ldaModel.score(X_test, y_test)
```

## Bagging Classifier

```python
from sklearn.ensemble import BaggingClassifier
BagClassModel = BaggingClassifier(
    base_estimator = DecisionTreeClassifier(),
    n_estimators = 100,
    max_samples = 0.8, 
    oob_score=True,
    random_state=5

)

BagClassModel.fit(X_train, y_train)
BagClassModel.oob_score_
```

```python
RandomModel = RandomForestClassifier(n_estimators = 100)
RandomModel.fit(X_train, y_train)
RandomModel.score(X_test, y_test)

from sklearn.ensemble import BaggingClassifier
BagClassModel = BaggingClassifier(
    base_estimator = RandomForestClassifier(n_estimators = 50),
    n_estimators = 100,
    max_samples = 0.8, 
    oob_score=True,
    random_state=5

)

BagClassModel.fit(X_train, y_train)
BagClassModel.oob_score_
```
