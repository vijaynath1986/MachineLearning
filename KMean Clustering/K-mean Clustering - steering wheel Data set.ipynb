```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
import random
```

```python
dat = []

for each in range(30):
    dat.append(random.randint(-20,-5))
    
for each in range(160):
    dat.append(random.randint(-5,5))
    
for each in range(20):
    dat.append(random.randint(5,20))
    
random.shuffle(dat)
```

```python
plt.scatter(dat, range(210))
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
```

```python
km_predicted = km.fit_predict(pd.DataFrame(dat))
```

```python
sns.scatterplot(dat, range(210), hue=km_predicted)
```

```python
km.predict(pd.DataFrame([[-3]]))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_dat = scaler.fit_transform(pd.DataFrame(dat))
scaled_dat = np.reshape(scaled_dat, 210)
```

```python
km_predicted_new = km.fit_predict(scaled_dat)
```

```python
sns.scatterplot(scaled_dat, range(210), hue=km_predicted_new, palette= ["red", "green", "blue"])
```
