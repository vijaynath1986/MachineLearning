```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./dataset/Mall_Customers.csv")
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.scatterplot(df, x = "Annual Income (k$)",  y="Spending Score (1-100)", hue="Gender", palette="deep")
```

```python
from sklearn.cluster import KMeans
```

### Lets group customers based on "Age" and "spending score", using K-mean clustering 

```python
x1 = df[["Age", "Spending Score (1-100)"]]
# WCSS - Within-Cluster Sum of Square
wcss = []

for each in range(1, 11):
    kmobj = KMeans(n_clusters=each)
    kmobj.fit_predict(x1)
    wcss.append(kmobj.inertia_)

print(wcss)
```

```python
plt.plot(wcss, 'ro-')
plt.xlabel("clusters")
plt.ylabel("Sum of sq disctance between clusters")
```

```python
kmobj = KMeans(n_clusters=4)
kmobj.fit_predict(x1)
kmobj.labels_
```

```python
sns.scatterplot(data = df, x = "Age", y = "Spending Score (1-100)", hue=kmobj.labels_, palette = 'deep')
plt.scatter(x = kmobj.cluster_centers_[:,:1], y = kmobj.cluster_centers_[:,1:2], color='black')
```
