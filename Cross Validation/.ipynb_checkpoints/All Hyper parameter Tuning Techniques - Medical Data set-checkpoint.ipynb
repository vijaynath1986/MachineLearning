```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../DataSets/breast-cancer-data.csv")
data.head()
```

```python
data.diagnosis.value_counts()
```

```python
X = data.iloc[:,2:]
y = data.diagnosis
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state = 0)
```

```python
# Default classification and no parameters
from sklearn.tree import DecisionTreeClassifier
decTreeObj = DecisionTreeClassifier()
decTreeObj.fit(train_X, train_y)
decTreeObj.score(test_X, test_y)
```

```python
#default parameters used by Decision Tree Classifier
decTreeObj.get_params()
```

 ## Grid Search CV method of hyper parameter tuning

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Select the classifier for hyper parameter tuning
decTreeObj = DecisionTreeClassifier()
# Give the possible input parameters to evaluate and run the classifier with those parameters
grid_obj = GridSearchCV(decTreeObj, param_grid= {
    'ccp_alpha': [0.1, 0.3, 0],
    'max_depth':[5, 10, 20],
    'min_samples_leaf': [1, 3, 5],}, cv = 10, return_train_score=False)
# Give train dataset and run all the classifer with different parameters
grid_obj.fit(train_X, train_y)
#view all the results in dataframe formate
pd.DataFrame(grid_obj.cv_results_)
```

```python
# Scorer function used on the "held out data" to choose the best parameters for the model.
grid_obj.score(test_X, test_y)
```

```python
# Estimator that was chosen by the search, i.e. estimator which gave highest score (or smallest loss if specified)
print(grid_obj.best_estimator_) 

# Mean cross-validated score of the best_estimator
print(grid_obj.best_score_)

## Parameter setting that gave the best results on the hold out data.
print(grid_obj.best_params_)

## The index (of the cv_results_ arrays) which corresponds to the best candidate parameter setting.
print(grid_obj.best_index_)
```

```python
# Using the above best_estimator to prediction
decTreeObj1 = DecisionTreeClassifier(max_depth=5, min_samples_leaf=5)
decTreeObj1.fit(train_X, train_y)
```

```python
decTreeObj1.predict(test_X)
decTreeObj1.score(test_X, test_y )
```

```python
grid_obj.best_score_
```

```python
df[["params", "mean_test_score"]]
```

```python
df.sort_values(by = "rank_test_score")
```

## Random Search Cross Validation method of hyper parameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Select the classifier for hyper parameter tuning
decTreeObj = DecisionTreeClassifier()
# Give the possible input parameters to evaluate and run the classifier with those parameters
random_obj = RandomizedSearchCV(decTreeObj, param_distributions= {
    'max_depth':[5, 10, 20],
    'min_samples_leaf': [1, 3, 5],}, cv = 5, return_train_score=False, n_iter=5)
# Give train dataset and run all the classifer with different parameters
random_obj.fit(train_X, train_y)
#view all the results in dataframe format
pd.DataFrame(random_obj.cv_results_)
```
