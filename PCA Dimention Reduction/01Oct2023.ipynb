```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv("../DataSets/trans_us.csv", thousands=',')
df.head()
```

```python
df.info()
```

```python
for val in df["1990"]:
    print(val)
```

```python
df["prevalence"] = df["TB prevalence, all forms (per 100 000 population per year)"]
```

```python
df
```

```python
df.drop(["TB prevalence, all forms (per 100 000 population per year)"], axis = 1, inplace= True)
```

```python
df
```

```python
X = df.drop(["prevalence"], axis=1)
```

```python
X
```

```python
from sklearn.decomposition import PCA
```

```python
pcaobj = PCA(n_components=3)
```

```python
pcaobj.fit(X)
```

```python
pcaobj.explained_variance_ratio_ * 100
```

```python
sns.lineplot(pcaobj.explained_variance_ratio_ * 100)
sns.scatterplot(pcaobj.explained_variance_ratio_ * 100, legend='full' )
```

```python
import matplotlib.pyplot as plt
```
